## **Loading the Dataset and selecting the required 42 input variables for our proposed model**



```python
dataset = pd.read_csv('/content/drive/MyDrive/Thesis/Dengue/DENGUE/Application/FDA APPROved drugs/FDA_approved_drugs_dataset_of_Only_Descriptors.csv')
dataset
```

## **Selecting the input variables for our proposed model and preprocessing the input dataset to use in our proposed model**

```python
predictors = dataset[["SpMax7_Bhi", "IC5", "nRotB", "SpMin8_Bhs", "SpAD_Dzv", "AATSC2e", "ATSC1v", "Kier2", "McGowan_Volume", "GATS1m", 
              "SHBint3", "SpMAD_Dze", "AATS2v", "CIC5", "VE2_Dzp", "VP-1", "BCUTc-1l", "SpMax_Dzp", "SHBint10", "minHBint7",
              "AATSC2c", "AATS1i", "maxwHBa", "minHBd", "GATS2i", "ETA_Shape_Y", "SHBint7", "VR2_D", "VR1_Dt", "VR3_Dzp", 
              "XLogP", "GATS5i", "BCUTp-1l", "ETA_BetaP_s", "AATS8i", "C1SP2", "GATS8s", "ATSC2c", "AATS4i", "AATS2i", 
              "VR1_Dzp", "minHBint9"]]


predictors.replace([np.inf, -np.inf], np.nan, inplace=True)

predictors = predictors.astype("float64")
predictors = pd.DataFrame(predictors).fillna(0)
```

## **The associated github repository has the file "finalized_model_ET_42.pkl" uploaded to it. It must be downloaded to your disk/drive.**

Steps: 
1.   Download the "finalized_model_ET_42.pkl" file to your disk/drive.
2.   Replace the URL in the DATA_PATH variable with the location where you placed the 'finalized_model_ET_42.pkl'

```python
import pickle

DATA_PATH = "/content/drive/MyDrive/Thesis/Dengue/DENGUE/Application"
infile = open(DATA_PATH+'/finalized_model_ET_42.pkl','rb')
loaded_model = pickle.load(infile)

#predict
pIC50 = loaded_model.predict(predictors)

predictors['pIC50'] = pIC50

output_dataset = dataset[['Name']]
output_dataset['pIC50'] = pIC50

output_dataset.to_csv('output_dataset.csv', index=False)

#download output file
from google.colab import files
files.download('output_dataset.csv') 
```
